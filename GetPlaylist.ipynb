```python
def check_transition_properties(M):
    #(1) a+b+c+d=0
    #(2) a^2+b^2+c^2+d^2 = 2
    #(3) |a|+|b|+|c|+|d| = 2
    # <--> (a+b+c+d)^2 + (a^2+b^2+c^2+d^2 - 2)^2 + (|a|+|b|+|c|+|d| - 2)^2 = 0
    row_sum = M.sum(1)
    row_abs_sum = np.abs(M).sum(1)
    row_sum_sq = np.sum(M**2,axis=1)
    pr = row_sum**2+(row_abs_sum-2)**2+(row_sum_sq-2)**2
    if pr.sum()==0:
        print('row properties satisfied')
        print('row sums:', row_sum)
        print('row abs sums:', row_abs_sum)
        print('row sum squares:', row_sum_sq)
    else:
        print('row properties are not satisfied')
```

```python
import numpy as np

def create_simplex_vertices(n):
    # Initialize the vertices with standard basis vectors
    V = np.eye(n)
    
    # Compute the additional vertex
    additional_vertex = (1/n) * (1 + np.sqrt(n+1)) * np.ones(n)
    V = np.vstack([V, additional_vertex])
    
    # Center the vertices at the origin
    centroid = np.mean(V, axis=0)
    V -= centroid
    
    # Normalize each vertex to have unit norm
    norms = np.linalg.norm(V, axis=1)
    V = V / norms[:, np.newaxis]
    
    return V

def generate_root_vectors(V):
    n = V.shape[0]  # Number of vertices
    d = V.shape[1] #dim
    # Initialize an empty list to store root vectors
    R = []
    
    # Generate all pairwise root vectors
    for i in range(n):
        for j in range(i + 1, n):
            root_vector = V[i] - V[j]
            R.append(root_vector)
            R.append(-root_vector)  # Include both directions for completeness
    
    # Convert list of root vectors to a numpy array
    R = np.array(R)
    s = 1+1/d
    #scale
    R /= s
    return R
```

```python
# Example usage
n = 3  # Dimension for a regular 3-simplex in R^3
V = create_simplex_vertices(n)

# print("Vertices of the regular", n, "-simplex in R^", n, ":\n", V)
print("VV^T:", V.dot(V.T))
print("Centroid of the vertices:",  np.mean(V, axis=0))
R = generate_root_vectors(V)

A = R.dot(V.T).round()
print(A)
check_transition_properties(A)
```

```python
def find_compatible_keys(key, level):
    # Define the Camelot wheel keys for minor and major separately
    minor_wheel = {
        "1A": "Ab minor",
        "2A": "Eb minor",
        "3A": "Bb minor",
        "4A": "F minor",
        "5A": "C minor",
        "6A": "G minor",
        "7A": "D minor",
        "8A": "A minor",
        "9A": "E minor",
        "10A": "B minor",
        "11A": "F# minor",
        "12A": "Db minor"
    }

    major_wheel = {
        "1B": "B major",
        "2B": "F# major",
        "3B": "Db major",
        "4B": "Ab major",
        "5B": "Eb major",
        "6B": "Bb major",
        "7B": "F major",
        "8B": "C major",
        "9B": "G major",
        "10B": "D major",
        "11B": "A major",
        "12B": "E major"
    }
    
    # Determine whether the key is major or minor
    if key.endswith("A"):
        wheel = minor_wheel
        other_wheel = major_wheel
    elif key.endswith("B"):
        wheel = major_wheel
        other_wheel = minor_wheel
    else:
        raise ValueError("Invalid key format. Please use Camelot wheel notation (e.g., '9B' for A minor)")

    # Get the given key's position in the wheel
    if key not in wheel:
        raise ValueError("Invalid key format. Please use Camelot wheel notation (e.g., '9B' for A minor)")
    
    index = list(wheel.keys()).index(key)
    compatible_keys = set()
    
    # Find directly adjacent keys and corresponding key from the other wheel
    adjacent_keys = [
        wheel[list(wheel.keys())[index - 1]],
        wheel[list(wheel.keys())[index + 1]],
    ]
    
    corresponding_key = other_wheel[list(other_wheel.keys())[index]]
    
    compatible_keys.update(adjacent_keys)
    compatible_keys.add(corresponding_key)
    
    return compatible_keys

# Example usage:
key = "2B"  # A minor
level = "direct"
compatible_keys = find_compatible_keys(key, level)
print(f"Keys compatible with {key} at '{level}' level: {compatible_keys}")
```
