```python
#Pink Floyd - https://open.spotify.com/artist/0k17h0D3J5VfsdmQ1iZtE9?si=61LGwos-R3KGikJIyyRFzA
#Neu! - https://open.spotify.com/artist/2CqEOngQOOMstJupNn6Hmf?si=SmPiuM6mRzuj3YGxMhTl-A
#Joy Division - https://open.spotify.com/artist/432R46LaYsJZV2Gmc4jUV5?si=2Xk4eD8FQKqTqrKvIwipQw
#Mingus - https://open.spotify.com/artist/1W8TbFzNS15VwsempfY12H?si=ro6QTQd1Rp2hLJEg0EXTfw
#Bowie - https://open.spotify.com/artist/0oSGxfWSnnOXhD2fKuz2Gy?si=YBjqb02jTC-ITxj8oC_i3A

artist_id = '0oSGxfWSnnOXhD2fKuz2Gy'
album_info = get_artist_albums(token, artist_id, offset=0, limit=None)
```

```python
album_name_ls = [a['name'] for a in album_info]
album_id_ls = [a['id'] for a in album_info]

print(album_name_ls)
print(album_id_ls)
tracks = [get_album_tracks(token, album_id, summarize=True) for album_id in album_id_ls]
tracks_df = pd.concat(tracks, axis=0)
tracks_df = tracks_df[tracks_df.artist_id == artist_id].reset_index(drop=True)
```

```python
#Dark side of the moon - https://open.spotify.com/album/4LH4d3cOWNNsVw41Gqt2kv?si=EaWNRXhmSx-BrhCGOiFZGQ
#Wish you were here - https://open.spotify.com/album/0bCAjiUamIFqKJsekOYuRw?si=Yo-P4KttSliS_PI87HrN4A
album_id = '0bCAjiUamIFqKJsekOYuRw'
album_tracks = get_album_tracks(token, album_id, summarize=True)
album_tracks.head()
```

```python
audio_feats = get_audio_features(token, tracks_df.track_id.tolist())
audio_feats_df = pd.DataFrame(audio_feats)
audio_feats_df = audio_feats_df.loc[:,['tempo', 'valence', 'liveness', 'instrumentalness',
       'acousticness', 'speechiness', 'mode', 'loudness', 'key', 'energy',
       'danceability']]
audio_feats_df
```

```python
na_mask = (audio_feats_df.isna().sum(1)==0).tolist()
audio_feats_df[na_mask]
```

```python
# compare_seeds_rec(seed_audio_feats_df, audio_feats_df[na_mask], include_categorical=True, circle_5 = True)

half_time=True
d_max = compare_seeds_rec(seed_audio_feats_df.copy(), audio_feats_df[na_mask].copy(), include_categorical=True, circle_5 = True)
if half_time:
    # seed_audio_feats_df.at[0, 'tempo'] = seed_audio_feats_df.at[0, 'tempo']* 2
    seed_audio_feats_df.at[0, 'tempo'] = seed_audio_feats_df.at[0, 'tempo']/ 2
    d_max_half = compare_seeds_rec(seed_audio_feats_df.copy(), audio_feats_df[na_mask].copy())
    # seed_audio_feats_df.at[0, 'tempo'] = seed_audio_feats_df.at[0, 'tempo']/ 2
    seed_audio_feats_df.at[0, 'tempo'] = seed_audio_feats_df.at[0, 'tempo']* 2
    
    d_max = np.vstack([d_max, d_max_half]).max(0)
    
print(np.sort(d_max)[::-1][:15])
print(np.argsort(d_max)[::-1][:15])

# rec_df.iloc[[0],:].copy()
top_sim_ix = np.argsort(d_max.flatten())[::-1][:15] #[1,None]
top_sims = d_max[top_sim_ix, None]


with pd.option_context('display.max_colwidth', None): 
    # display(rec_df.sort_values('release_date', ascending=True))
    display(tracks_df.iloc[top_sim_ix,:])
```
